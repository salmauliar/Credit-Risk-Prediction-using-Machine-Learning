```python
# =============================
# FINAL TASK – CREDIT RISK PREDICTION
# =============================

# 1. IMPORT LIBRARY
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import warnings
warnings.filterwarnings('ignore')

# 2. LOAD DATA
df = pd.read_csv('loan_data_2007_2014.csv', low_memory=False)
print("Shape awal:", df.shape)
print(df.info())

# Drop kolom dengan cardinality tinggi yang bikin memory meledak
cols_to_drop = ['id', 'member_id', 'emp_title', 'title', 'desc', 'url', 'zip_code']
df = df.drop(columns=[col for col in cols_to_drop if col in df.columns])

# 3. TARGET ENCODING
df = df[df['loan_status'].isin(['Fully Paid', 'Charged Off'])]
df['label'] = df['loan_status'].apply(lambda x: 1 if x == 'Fully Paid' else 0)
print(df['loan_status'].value_counts())

# 4. EDA (Exploratory Data Analysis)
# Distribusi interest rate
df['int_rate'].hist(bins=30)
plt.title('Distribusi Interest Rate')
plt.xlabel('Interest Rate')
plt.ylabel('Frekuensi')
plt.show()

# Korelasi antar fitur numerik
numerical_cols = df.select_dtypes(include='number').columns
plt.figure(figsize=(12, 10))
sns.heatmap(df[numerical_cols].corr(), cmap='coolwarm', annot=False)
plt.title('Heatmap Korelasi')
plt.show()

# Boxplot interest rate vs loan status
sns.boxplot(x='loan_status', y='int_rate', data=df)
plt.title('Interest Rate berdasarkan Status Pinjaman')
plt.show()

# 5. ENCODING KATEGORIKAL
df_encoded = df.copy()
categorical_cols = df_encoded.select_dtypes(include='object').columns.tolist()
if 'loan_status' in categorical_cols:
    categorical_cols.remove('loan_status')

le = LabelEncoder()
for col in categorical_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))

print("Kolom yang di-encode:", categorical_cols)
print(df_encoded[categorical_cols].head())

# 6. PISAHKAN FITUR DAN TARGET
X = df_encoded.drop(columns=['loan_status', 'label'])
y = df_encoded['label']

# 7. DATA CLEANING: Missing Value & Outlier → Mean
X = X.replace([np.inf, -np.inf], np.nan)
X = X.fillna(X.mean(numeric_only=True))

# Deteksi dan imputasi outlier
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    mean_val = X[col].mean()
    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    X[col] = np.where((X[col] < lower) | (X[col] > upper), mean_val, X[col])

# 8. NORMALISASI (Min-Max Scaling)
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# 9. SPLIT DATA
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print("Train shape:", X_train.shape)
print("Test shape :", X_test.shape)

# 10. LOGISTIC REGRESSION
# 1. Pastikan hanya numerik
X_train = X_train.select_dtypes(include=['int64', 'float64'])
X_test = X_test.select_dtypes(include=['int64', 'float64'])

# 2. Imputasi NaN 
X_train = X_train.fillna(0)
X_test = X_test.fillna(0)

# 3. Pastikan tidak ada NaN tersisa
print("✅ X_train NaN:", X_train.isnull().sum().sum())
print("✅ X_test NaN:", X_test.isnull().sum().sum())

logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
# Logistic Regression model
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
y_pred_log = logreg.predict(X_test)

print("=== Logistic Regression ===")
print(confusion_matrix(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))
print("ROC AUC:", roc_auc_score(y_test, logreg.predict_proba(X_test)[:, 1]))

# 11. DECISION TREE
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)

print("=== Decision Tree ===")
print(confusion_matrix(y_test, y_pred_tree))
print(classification_report(y_test, y_pred_tree))
print("ROC AUC:", roc_auc_score(y_test, tree.predict_proba(X_test)[:, 1]))

# 12. ROC CURVE
fpr_log, tpr_log, _ = roc_curve(y_test, logreg.predict_proba(X_test)[:, 1])
fpr_tree, tpr_tree, _ = roc_curve(y_test, tree.predict_proba(X_test)[:, 1])

plt.figure(figsize=(8,6))
plt.plot(fpr_log, tpr_log, label='Logistic Regression')
plt.plot(fpr_tree, tpr_tree, label='Decision Tree')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
# =====================================
# FINAL – K-MEANS CLUSTERING + VISUALISASI
# =====================================

# 1. IMPORT LIBRARY
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# 2. LOAD DATA (versi ringan)
selected_cols = [
    'loan_status', 'int_rate', 'loan_amnt', 'annual_inc',
    'term', 'grade', 'purpose', 'dti', 'open_acc',
    'revol_bal', 'revol_util', 'total_acc'
]

df = pd.read_csv('loan_data_2007_2014.csv', usecols=selected_cols, engine='python')

# 3. ENCODING KATEGORIKAL
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cat_cols = df.select_dtypes(include='object').columns.tolist()
if 'loan_status' in cat_cols:
    cat_cols.remove('loan_status')  # jika masih ada

for col in cat_cols:
    df[col] = le.fit_transform(df[col].astype(str))

# 4. HAPUS KOLOM NON-FEATURE
X = df.drop(columns=['loan_status'], errors='ignore')

# 5. BERSIHKAN DATA
X = X.replace([np.inf, -np.inf], np.nan)
X = X.fillna(X.mean(numeric_only=True))

# 6. NORMALISASI (Min-Max)
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# 7. TENTUKAN JUMLAH CLUSTER DENGAN ELBOW METHOD
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o')
plt.title('Elbow Method – Optimal Jumlah Cluster')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
# =============================
# FINAL TASK – CREDIT RISK PREDICTION
# =============================

# 1. IMPORT LIBRARY
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score


# 2. LOAD DATA
df = pd.read_csv('loan_data_2007_2014.csv', low_memory=False)
print("Shape awal:", df.shape)
print(df.info())

# Drop kolom dengan cardinality tinggi yang bikin memory meledak
cols_to_drop = ['id', 'member_id', 'emp_title', 'title', 'desc', 'url', 'zip_code']
df = df.drop(columns=[col for col in cols_to_drop if col in df.columns])

# 3. TARGET ENCODING
df = df[df['loan_status'].isin(['Fully Paid', 'Charged Off'])]
df['label'] = df['loan_status'].apply(lambda x: 1 if x == 'Fully Paid' else 0)
print(df['loan_status'].value_counts())

# 4. EDA (Exploratory Data Analysis)
# Distribusi interest rate
df['int_rate'].hist(bins=30)
plt.title('Distribusi Interest Rate')
plt.xlabel('Interest Rate')
plt.ylabel('Frekuensi')
plt.show()

# Korelasi antar fitur numerik
numerical_cols = df.select_dtypes(include='number').columns
plt.figure(figsize=(12, 10))
sns.heatmap(df[numerical_cols].corr(), cmap='coolwarm', annot=False)
plt.title('Heatmap Korelasi')
plt.show()

# Boxplot interest rate vs loan status
sns.boxplot(x='loan_status', y='int_rate', data=df)
plt.title('Interest Rate berdasarkan Status Pinjaman')
plt.show()

# 5. ENCODING KATEGORIKAL
df_encoded = df.copy()
categorical_cols = df_encoded.select_dtypes(include='object').columns.tolist()
if 'loan_status' in categorical_cols:
    categorical_cols.remove('loan_status')

le = LabelEncoder()
for col in categorical_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))

print("Kolom yang di-encode:", categorical_cols)
print(df_encoded[categorical_cols].head())

# 6. PISAHKAN FITUR DAN TARGET
X = df_encoded.drop(columns=['loan_status', 'label'])
y = df_encoded['label']

# 7. DATA CLEANING: Missing Value & Outlier → Mean
X = X.replace([np.inf, -np.inf], np.nan)
X = X.fillna(X.mean(numeric_only=True))

# Deteksi dan imputasi outlier
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    mean_val = X[col].mean()
    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    X[col] = np.where((X[col] < lower) | (X[col] > upper), mean_val, X[col])

# 8. NORMALISASI (Min-Max Scaling)
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# 9. SPLIT DATA
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print("Train shape:", X_train.shape)
print("Test shape :", X_test.shape)

# 10. LOGISTIC REGRESSION
# 1. Pastikan hanya numerik
X_train = X_train.select_dtypes(include=['int64', 'float64'])
X_test = X_test.select_dtypes(include=['int64', 'float64'])

# 2. Imputasi NaN 
X_train = X_train.fillna(0)
X_test = X_test.fillna(0)

# 3. Pastikan tidak ada NaN tersisa
print("✅ X_train NaN:", X_train.isnull().sum().sum())
print("✅ X_test NaN:", X_test.isnull().sum().sum())

# Tuning parameter Logistic Regression
param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear'],  # 'liblinear' support l1 dan l2
    'class_weight': [None, 'balanced'],
    'max_iter': [500, 1000]
}

grid_search = GridSearchCV(
    estimator=LogisticRegression(),
    param_grid=param_grid,
    scoring='roc_auc',  # pakai ROC AUC karena dataset tidak selalu seimbang
    cv=5,
    n_jobs=-1,
    verbose=1
)

# Fit grid search
grid_search.fit(X_train, y_train)

# Model terbaik
best_logreg = grid_search.best_estimator_

# Prediksi & evaluasi
y_pred_log = best_logreg.predict(X_test)
y_proba_log = best_logreg.predict_proba(X_test)[:, 1]

print("=== Best Parameters ===")
print(grid_search.best_params_)

print("\n=== Logistic Regression (Tuned) ===")
print(confusion_matrix(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))
print("ROC AUC:", roc_auc_score(y_test, y_proba_log))


# 11. DECISION TREE
# Siapkan parameter grid
param_grid_tree = {
    'max_depth': [3, 5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

grid_tree = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_grid=param_grid_tree,
    scoring='roc_auc',
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_tree.fit(X_train, y_train)
best_tree = grid_tree.best_estimator_

# Prediksi & evaluasi
y_pred_tree = best_tree.predict(X_test)
y_proba_tree = best_tree.predict_proba(X_test)[:, 1]

print("=== Best Parameters - Decision Tree ===")
print(grid_tree.best_params_)

print("\n=== Decision Tree (Tuned) ===")
print(confusion_matrix(y_test, y_pred_tree))
print(classification_report(y_test, y_pred_tree))
print("ROC AUC:", roc_auc_score(y_test, y_proba_tree))

# ===============================
# TABEL PERBANDINGAN MODEL
# ===============================

model_comparison = pd.DataFrame({
    'Model': ['Logistic Regression (Tuned)', 'Decision Tree (Tuned)'],
    'Accuracy': [
        accuracy_score(y_test, y_pred_log),
        accuracy_score(y_test, y_pred_tree)
    ],
    'Precision': [
        precision_score(y_test, y_pred_log),
        precision_score(y_test, y_pred_tree)
    ],
    'Recall': [
        recall_score(y_test, y_pred_log),
        recall_score(y_test, y_pred_tree)
    ],
    'F1 Score': [
        f1_score(y_test, y_pred_log),
        f1_score(y_test, y_pred_tree)
    ],
    'ROC AUC': [
        roc_auc_score(y_test, y_proba_log),
        roc_auc_score(y_test, y_proba_tree)
    ]
})

print("\n=== TABEL PERBANDINGAN MODEL ===")
print(model_comparison)

# 12. ROC CURVE
fpr_log, tpr_log, _ = roc_curve(y_test, y_proba_log)
fpr_tree, tpr_tree, _ = roc_curve(y_test, y_proba_tree)

# ROC Curve update
fpr_log, tpr_log, _ = roc_curve(y_test, y_proba_log)
fpr_tree, tpr_tree, _ = roc_curve(y_test, y_proba_tree)

plt.figure(figsize=(8, 6))
plt.plot(fpr_log, tpr_log, label='Logistic Regression (Tuned)')
plt.plot(fpr_tree, tpr_tree, label='Decision Tree (Tuned)')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve – Logistic vs Decision Tree')
plt.legend()
plt.show()
```
